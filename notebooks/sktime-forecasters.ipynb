# VAR FORECASTING

```python
from sktime.forecasting.var import VAR
from sktime.forecasting.base import ForecastingHorizon
```

```python
var_forecaster = VAR(freq = 'M')
```

```python
train = data.loc[(data.index>'2011-12')&(data.index< '2015-02')]
validation = data.loc[data.index >= '2015-02']
```

```python
# fh = np.arange(1,12)
```

```python
fh = ForecastingHorizon(
    pd.PeriodIndex(pd.date_range("2015-02", periods= 12, freq="M")), is_relative=False
)
```

```python
fh
```

```python
var_forecaster =  var_forecaster.fit(train, fh = fh)
```

```python
var_pred = var_forecaster.predict()
```

```python
var_pred
```

Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
valid_period = validation[:'2016-01']
```

```python
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(valid_period, var_pred, symmetric = False)
```

```python
mape
```

```python
r2_var  = r2_score(valid_period, var_pred)
```

```python
r2_var
```

```python
multindex = multindex.drop("year", axis = 1)
```

```python
hier_data = multindex.groupby(["Warehouse", "Product_Category","Product_Code", "month_year"], as_index=False)['Order_Demand'].sum()
```

```python
hier_data.head()
```

```python
category_combination = multindex[['Warehouse', 'Product_Category', 'Product_Code']].drop_duplicates()
```

```python
time = pd.DataFrame(hier_data.month_year.drop_duplicates())
```

```python
basis = category_combination.merge(time, how='cross')
```

```python
new_df = basis.merge( hier_data, on=['Warehouse', 'Product_Category', 'Product_Code', 'month_year'], how='left')
```

```python
new_df = new_df.fillna(1.0)
```

```python
new_df["month_year"] = pd.to_datetime(new_df["month_year"], format = '%Y-%m')
new_df["month_year"] = new_df["month_year"].dt.to_period('M')
```

```python
new_df = new_df[new_df['month_year'] >= "2012-01"]
```

```python
new_df = new_df.groupby(['Warehouse', 'Product_Category', 'Product_Code', 'month_year'])["Order_Demand"].sum()
```

```python
new_df = new_df.to_frame()
```

```python
new_df.head()
```

```python
idx = pd.IndexSlice
train_hier = new_df[new_df.index.get_level_values('month_year') <= "2015-01"]
```

```python
validation_hier = new_df[new_df.index.get_level_values('month_year')>"2015-01"]
```

```python
from sktime.registry import all_estimators
hier_forecasters = all_estimators(
    "forecaster", as_dataframe=True, return_tags=["scitype:y", "y_inner_mtype"]
)
```

```python
# Ensemble Forecaster
from sktime.forecasting.compose import EnsembleForecaster
from sktime.forecasting.naive import NaiveForecaster
from sktime.forecasting.trend import PolynomialTrendForecaster

forecasters = [
    ("trend", PolynomialTrendForecaster()),
    ("naive", NaiveForecaster()),
]
forecaster = EnsembleForecaster(forecasters=forecasters, weights=[4, 10])
```

```python
forecaster = forecaster.fit(y = train_hier, fh=fh)
```

```python
pred = forecaster.predict()
```

```python
pred
```

```python
validate = validation_hier[validation_hier.index.get_level_values('month_year')<"2016-02"]
```

```python
score_ensemble = r2_score(validate, pred)
```

```python
score_ensemble
```
